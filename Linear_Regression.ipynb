```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
```

```python
life_satisfaction = pd.read_csv("./Data Sets/life_satisfaction.csv")
life_satisfaction.head()
```

```python
X = life_satisfaction['GDP per capita (USD)'].values # Feature
Y = life_satisfaction['Life satisfaction'].values # Target
```

```python
life_satisfaction.plot(kind='scatter', x='GDP per capita (USD)', y='Life satisfaction')
plt.axis([25_000, 60_000, 5, 8]) # [xmin, xmax, ymin, ymax]
plt.show() # GDP per capita vs Life satisfaction
```

```python
linearRegression = LinearRegression()
linearRegression.fit(X.reshape(-1, 1), Y) # Reshape X to 2D array
```

```python
kNeighborsRegressor = KNeighborsRegressor(n_neighbors=3)
kNeighborsRegressor.fit(X.reshape(-1, 1), Y) # Reshape X to 2D array
```

```python
X_new = [[37_655.2]] # Cyprus GDP per capita
print("Life Satisfaction index is : ", linearRegression.predict(X_new)) # Predict Cyprus life satisfaction
print("Life Satisfaction index is : ", kNeighborsRegressor.predict(X_new))
```
